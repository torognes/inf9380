```python
a = np.zeros((4,4))
a[1:3, 1:3] = 3
pd.DataFrame(a).style
```

```python
a = np.zeros((4,4))
a[1:3, 1:3] = 3

b = a[1:3,1:]
a[2,2] = 1234
pd.DataFrame(b).style
#pd.DataFrame(a).style
```

```python
a = np.zeros((4,4))
a[1:3, 1:3] = 3

b = a[1:3,1:].copy()
a[2,2] = 1234
pd.DataFrame(b).style
#pd.DataFrame(a).style
```

```python
a = np.zeros((4,4))
a[1:3, 1:3] = 3

b = a[1:3,1:]
#a[2,2] = 1234
#pd.DataFrame(b).style
b[0,0] = 111
pd.DataFrame(a).style
```

## Array manipulation

```python
mat = np.array([[1,2,3], [4,5,6]])
mat
```

```python
x = mat.reshape(3,2)
x
```

```python
mat.ravel()
```

```python
mat1 = np.array([[1,2,3],[4,5,6]])
mat2 = np.array([[7,8,9],[10,11,12]])
mat3 = np.array([[1,1,1],[2,2,2]])
x = np.concatenate((mat1,mat2,mat3))
x
```

```python
x[1,2]
```

```python
a = np.split(x, 3, axis=1)
pd.DataFrame(a[2]).style
```
